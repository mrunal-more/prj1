```python
import pandas as pd

from sklearn.preprocessing import LabelEncoder
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import MinMaxScaler, StandardScaler

import pickle 
import json 
```

```python
data = pd.read_csv('age_of_marriage_data.csv')
print(data.shape)
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.shape
```

```python
data.head(2)
```

```python
x = data.loc[:,['gender','height','religion','caste','mother_tongue','country']]
y = data.age_of_marriage
```

```python
x
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
x.loc[:,['gender','religion','caste','mother_tongue','country']]= \
x.loc[:,['gender','religion','caste','mother_tongue','country']].apply(enc.fit_transform)
```

```python
x
```

```python
x.height.unique()
```

```python
int(x.loc[1,'height'].split('\'')[0])*30.48 ##feet to cm
```

```python
int(x.loc[1,'height'].split('\'')[1].replace('"',''))*2.54 ##inches to cm
```

```python
def h_cms(h):
    return int(h.split('\'')[0])*30.48+\
    int(h.split('\'')[1].replace('"',''))*2.54
```

```python
x['height_cms'] = x.height.apply(h_cms)
```

```python
x.head()
```

```python
x.drop('height',inplace=True,axis=1)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=101)
```

```python
linear_model = LinearRegression() # GDA
linear_model.fit(x_train,y_train)
```

```python
train_score = linear_model.score(x_train,y_train)
print(train_score)

test_score = linear_model.score(x_test,y_test)
print(test_score)
```

```python
knn_model = KNeighborsRegressor()
knn_model.fit(x_train,y_train)  ## lazy learner
```

```python
train_score = knn_model.score(x_train,y_train)
print(train_score)

test_score = knn_model.score(x_test,y_test)
print(test_score)
```

```python
scaler = MinMaxScaler()
arr = scaler.fit_transform(x)
x_scale = pd.DataFrame(arr, columns = x.columns)
x_scale
```

```python
x_scale_train,x_scale_test, y_train, y_test = train_test_split(x_scale,y, test_size=0.2, random_state=17)
```

```python
knn_model = KNeighborsRegressor()
knn_model.fit(x_scale_train,y_train)  ## lazy learner

train_score = knn_model.score(x_scale_train,y_train)
print(train_score)

test_score = knn_model.score(x_scale_test,y_test)
print(test_score)
```

```python
with open('model.pkl','wb') as file: 
    pickle.dump(knn_model,file)
```

```python
gender="male"
religion="Hindu"
caste="Brahmin"
mother_tongue="Marathi"
country="India"
height_cms=157.48



if gender == "male":
    gender = 1
else: 
    gender = 0
    
    
if religion=="Hindu":
    religion=1
elif religion=="Christian":
    religion=2
elif religion=="Jain":
    religion=3
elif religion=="Sikh":
    religion=4
else:
    religion=5
    
    
if caste=="Shwetamber":
    caste=1
elif caste=="Brahmin":
    caste=2
elif caste=="Thakur":
    caste=3
else:
    caste=4
    
if mother_tongue=="Marathi":
    mother_tongue=1
elif mother_tongue=="Hindi":
    mother_tongue=2
elif mother_tongue=="Gujrati":
    mother_tongue=3
elif mother_tongue=="English":
    mother_tongue=4
else:
    mother_tongue=5
    
    
if country=="India":
    country=1
elif country=="USA":
    country=2
elif country=="Canada":
    country=3
else:
    country=4
    

user_data = np.zeros(len(x.columns))
user_data[0] = gender
user_data[1] = religion
user_data[2] = caste
user_data[3] = mother_tongue
user_data[2] = country
user_data[3] = height_cms      
```

```python

### Scaling the user data 
user_data_scale = scaler.transform([user_data])
print(user_data_scale)

result = knn_model.predict(user_data_scale)
result
```
